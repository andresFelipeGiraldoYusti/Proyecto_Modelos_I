# Funciones:

```python
def contar(df, columna): # Cuenta la frecuencia de cada valor único en una columna
  return df[columna].value_counts()

def filtrar(df, columna, valor): # Filtra un valor de una columna
  return df[df[columna] == valor]

def contar_rango(df,columna, liminf, limsup):
  return df[(df[columna] >= liminf) & (df[columna] <= limsup)].shape[0]
```
